```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(rc={'figure.figsize':(10,8)})
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# grafico simple

# datos
x = df['month_number']
y=df['total_profit']

# grafico

# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar
plt.plot(
    x, # eje x
    y, # eje y
    label='linea', # etiquetado
    color="purple", # color
    linewidth=1 # tamano de la curva
)
#plt.legend() # agregar etiquetado
plt.title("Total profit per month") # agregar titulo
plt.xlabel('Month') # nombre eje x
plt.ylabel('Total Profit') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

```python
# datos
x = df['month_number']
y=df['total_profit']


# grafico
fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    x, # eje x
    y, # eje y
    #s, # tamano de los puntos
    c="purple", # color
    alpha=0.7, # intensidad color
    marker='o', # forma de los puntos
    label="Total profit per month" # etiquetdo fijando posicion
)
plt.xlabel("Month")
plt.ylabel("Total profit")
plt.legend(loc='upper left')
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
# grafico compuesto

# datos
y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5 = df['shampoo']
y6 = df['moisturizer']

x = df['month_number']

# grafico
# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar

# a) facecream
plt.plot(
    x, # eje x
    y1, # eje y
    label='facecream', # etiquetado
    color="black", # color
    linewidth=1 # tamano de la curva
)
# b) facewash
plt.plot(
    x, # eje x
    y2, # eje y
    label='facewash', # etiquetado
    color="b", # color
    linewidth=1 # tamano de la curva
)
# c) toothpaste 
plt.plot(
    x, # eje x
    y3, # eje y
    label='toothpaste', # etiquetado
    color="r", # color
    linewidth=1 # tamano de la curva
)
# d) bathingsoap 
plt.plot(
    x, # eje x
    y4, # eje y
    label='bathingsoap', # etiquetado
    color="purple", # color
    linewidth=1 # tamano de la curva    
)    
# e) shampoo 
plt.plot(
    x, # eje x
    y5, # eje y
    label='shampoo ', # etiquetado
    color="orange", # color
    linewidth=1 # tamano de la curva 
)    
# f) moisturizer 
plt.plot(
    x, # eje x
    y6, # eje y
    label='moisturizer', # etiquetado
    color="g", # color
    linewidth=1 # tamano de la curva    
    
)

plt.legend() # agregar etiquetado
plt.title("Ventas por producto") # agregar titulo
plt.xlabel('Month') # nombre eje x
plt.ylabel('Ventas') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# datos

y1 = df['facecream']
y2 = df['facewash']

x = df['month_number']

# grafico
# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar

# a) facecream
plt.bar(
    x, # eje x
    y1, # eje y
    label='facecream', # etiquetado
    align='center', # centrar nombre eje x
    color="black", # color
    linewidth=1 # tamano de la curva
)
# b) facewash
plt.bar(
    x, # eje x
    y2, # eje y
    label='facewash', # etiquetado
    align='center', # centrar nombre eje x
    color="b", # color
    linewidth=1 # tamano de la curva
)

plt.xlabel('Month', size=25)
plt.ylabel('Ventas por mes',size=25)
plt.legend(loc='upper left')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
#datos
datos = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
sns.set(style="whitegrid")

#figura
fig = plt.figure(figsize =(10, 7))
 
#gráfico
sns.boxplot(data=df[datos])
 
# Se muestra el gráfico
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
total = df['total_units'].sum()
print(total)
nombre=list(df.columns.values[1:-2])
data=df[nombre].sum()
cmap = plt.cm.Set3
colors = cmap(np.linspace(0., 1., 9))

fig, ax=plt.subplots()
g = ax.pie(
    data,
    labels=nombre,
    colors=colors,
    autopct='%1.1f%%',
)
plt.title('Total= '+str(total))
plt.show()
```
